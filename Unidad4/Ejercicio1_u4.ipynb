```python
import scipy.stats as stats

# Datos proporcionados
media = 70  # kilogramos
desviacion_estandar = 4  # kilogramos
total_estudiantes = 500

# Rango de peso
peso_inferior = 68
peso_superior = 72

# Calcular la probabilidad de que un estudiante pese menos de 72 kg
prob_menos_72 = stats.norm.cdf(peso_superior, loc=media, scale=desviacion_estandar)

# Calcular la probabilidad de que un estudiante pese menos de 68 kg
prob_menos_68 = stats.norm.cdf(peso_inferior, loc=media, scale=desviacion_estandar)

# La probabilidad de estudiantes que pesan más de 68 y menos de 72 kg
probabilidad_rango = prob_menos_72 - prob_menos_68

# Cuantos estudiantes pesan más de 68 y menos de 72 kg
num_estudiantes_rango = probabilidad_rango * total_estudiantes

print(f"La probabilidad de estudiantes que pesan más de {peso_inferior} y menos de {peso_superior} kg es: {probabilidad_rango:.4f}")
print(f"El número de estudiantes que pesan más de {peso_inferior} y menos de {peso_superior} kg es: {num_estudiantes_rango:.2f}")
```
