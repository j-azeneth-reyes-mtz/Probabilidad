```python
import matplotlib.pyplot as plt
import math
```

```python
# Valores de la variable aleatoria
x = [0, 10, 20, 50]
p = [0.50, 0.30, 0.15, 0.05]
```

```python
# Verificar suma = 1
print("Suma de probabilidades:", sum(p))
```

```python
#Calcular E(x)
esperanza=0
for i in range(len(x)):
  esperanza += x[i] * p[i]

#Calcular E(X elevado al cuadrado)
esperanza_cuadrado = 0
for i in range(len(x)):
   esperanza_cuadrado += (x[i] ** 2) * p[i]
```

```python
import math
#Calcular varianza y desviacion estandar (moved outside the loop)
varianza = esperanza_cuadrado - (esperanza ** 2)
desviacion = math.sqrt(varianza)
#Mostrar resultados (moved outside the loop)
print(f"E(X)= {esperanza:.2F}")

print(f"Varianza= {varianza:.2F}")
print(f"Desviacion estandar= {desviacion:.2F}")
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
#Calcular funcion acumulada
F=np.cumsum(p)
#Mostrar resutados
for xi, fi, in zip(x, F):
  print(f"F({xi}) = {fi:.2f}")

#Graficar
plt.step(x, F, where='post')
plt.title('Funcion de distribucion acumulada')
plt.xlabel('Calificacion')
plt.ylabel('Probabilidad acumulada')
plt.grid(True)
plt.ylim(0, 1.05)
plt.show()
```
