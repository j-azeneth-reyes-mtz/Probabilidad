```python
import matplotlib.pyplot as plt
import math
```

```python
# Valores de la variable aleatoria
x = [0, 1, 2, 3, 4, 5]
p = [0.10, 0.25, 0.30, 0.20, 0.10, 0.05]
```

```python
# Verificar suma = 1
print("Suma de probabilidades:", sum(p))
```

```python
valor_esperado= 0
```

```python
#Recorrer los valores y sus probabilidades
for i in range(len(x)):
  xi=x[i] # Valor de la variable aleatoria
  pi=p[i] # Probabilidad de ese valor
  producto=xi*pi #Multiplicar valor por su probabilidad
  valor_esperado +=producto
  #Mostrar cada paso
  print(f"x = {xi}, p = {pi}, xi*pi = {producto:.2f}")

  #Mostrar resultado final
  print(f"\nValor esperado (Esperanza): {valor_esperado:.2f}")
```

```python
#Calcular E(x)
esperanza=0
for i in range(len(x)):
  esperanza += x[i] * p[i]

#Calcular E(X elevado al cuadrado)
esperanza_cuadrado = 0
for i in range(len(x)):
   esperanza_cuadrado += (x[i] ** 2) * p[i]
```

```python
import math
#Calcular varianza y desviacion estandar (moved outside the loop)
varianza = esperanza_cuadrado - (esperanza ** 2)
desviacion = math.sqrt(varianza)
#Mostrar resultados (moved outside the loop)
print(f"E(X)= {esperanza:.2F}")
print(f"E(X^2)= {esperanza_cuadrado:.2F}")
print(f"Varianza= {varianza:.2F}")
print(f"Desviacion estandar= {desviacion:.2F}")
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
#Calcular funcion acumulada
F=np.cumsum(p)
#Mostrar resutados
for xi, fi, in zip(x, F):
  print(f"F({xi}) = {fi:.2f}")

  #Graficar
  plt.step(x, F, where='post')
  plt.title('Funcion de distribucion acumulada')
  plt.xlabel('Calificacion')
  plt.ylabel('Probabilidad acumulada')
  plt.grid(True)
  plt.ylim(0, 1.05)
  plt.show()
```
