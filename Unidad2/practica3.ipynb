```python
import math

# 1. Existen diez asistentes de profesor disponibles para calificar exámenes en un curso de cálculo en una gran universidad.
# 2. El primer examen se compone de cuatro preguntas y el profesor desea seleccionar un asistente diferente para calificar cada pregunta (sólo un asistente por pregunta).
# 3. ¿De cuántas maneras se pueden elegir los asistentes para calificar?
# 4. En este caso n = tamaño del grupo = 10 y k = tamaño del subconjunto = 4.

# Cálculo del número de permutaciones:
# n = 10 (total de asistentes)
# k = 4 (número de preguntas/asistentes a seleccionar)

n = 10
k = 4

# La fórmula para permutaciones es P(n, k) = n! / (n-k)!
num_permutations = math.factorial(n) / math.factorial(n - k)

print(f"El número de asistentes disponibles (n): {n}")
print(f"El número de preguntas (k): {k}")
print(f"El número de maneras en que se pueden elegir los asistentes para calificar es: {int(num_permutations)}")
# El número de permutaciones es 10! / (10-4)! = 10! / 6! = 3,628,800 / 720 = 5040
```
