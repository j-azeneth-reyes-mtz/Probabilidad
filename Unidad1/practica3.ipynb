```python
import random
edades = []
for i in range(100):
  numero_aleatorio = random.randint(18, 70)
  edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
#crear un DataFrame de pandas con los numeros aleatorios
df = pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
#Calcular la tabla de distribucion de frecuencias
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

#Imprimir la tabla de distribucion de frecuencias
print("Tabla de Distribucion de Frecuencias:")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

```python
# Importamos  la libreria math de funciones matematicas
import math
```

```python
# Volvemos a usar la lista edades
print(edades)
```

```python
#crear un DataFrame con las edades
df = pd.DataFrame({'Edad': edades})
df
```

```python
#calcular el número de intervalos utilizando la regla de Sturges
n = len(edades)
K = int( 1 + math.log2(n))
print("Valores de n = ", n , "\nNúmero de intervalos =", K)
```

```python
#Crear la tabla de frecuencias
tabla_frecuencias = pd.cut(df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencias.columns = ['Clase', 'Frec']
tabla_frecuencias
```

```python
#Ordenar la tabla de frecuencia por el intervalo
tabla_frecuencias = tabla_frecuencias.sort_values(by='Clase')
tabla_frecuencias
```

```python
#Generar columnas siguientes
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frec'] / n
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frec'] / n * 100
tabla_frecuencias['Frec Acumulada'] = tabla_frecuencias['Frec'].cumsum()
tabla_frecuencias['Frec Relativa Acumulada'] = tabla_frecuencias['Frec Relativa'].cumsum()
tabla_frecuencias['Frec Porcentual Acumulada'] = tabla_frecuencias['Frec Porcentual'].cumsum()

#Mostrar la tabla de frecuencia
print(tabla_frecuencias)
```
