```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
covid = pd.read_csv("/content/20-06-30-COVID19MEXICO - codificado.csv" , encoding="ISO-8859-1")
```

```python
covid.head()
```

```python
covid.info()
```

```python
#Mostrar la edad maxima de la base de datos de la columna edad
covid.EDAD.max()
```

```python
#Mostrar la edad minima de la base de datos de la columna edad
covid.EDAD.min()
```

```python
#Dibuja el histograma del atributo "EDAD" con bins = 3
#Calcular la media y guardarla en una variable para mostrarla en el histograma
media = covid.EDAD.mean()
plt.hist(covid.EDAD, bins = 11)
plt.title("Histograma Edad Pruebas COVID")
plt.axvline(media, color='y', linestyle='dashed', linewidth=2, label=f'Media{media:.2f}')
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
```

```python
#Dibuja el histograma de la columna sexo,
#esto permite comparar cuantos hay de cada grupo
plt.hist(covid.SEXO, bins = 3)

#establece las etiquetas x/y y muestre el título
plt.title("Histograma Sexo Pruebas COVID")
plt.xlabel("Sexo")
plt.ylabel("Frecuencia")
```

```python
#Dibuja el histograma del atributo "Tipo Paciente", esto permite
#comparar cuantos hay de cada grupo
plt.hist(covid.TIPO_PACIENTE, bins = 3)

#establece las etiquetas x/y y muestra el título
plt.title("Histograma de tipos Pacientes en Pruebas COVID")
plt.xlabel("Tipo de Paciente")
plt.ylabel("Frecuencia")
```

```python
#Crea un diagrama de cajas para visulizar la distribucion
#de los rangos de edad por sexo
sns.boxplot(x = "SEXO", y = "EDAD", data=covid)
```

```python
#Crear un diagrama de cajas para visualizar la distribucion de
#los rangos de edad por tipo de paciente
sns.boxplot(x = "TIPO_PACIENTE", y = "EDAD", data = covid)
```
