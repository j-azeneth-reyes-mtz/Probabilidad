```python
import random
import matplotlib.pyplot as plt
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
  numero_aleatorio = round(random.uniform(50, 100), 2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  numero_aleatorio=round(random.uniform(50, 100), 2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
maximo1 = max(fabrica1)
minimo1 = min(fabrica1)
rango1 = maximo1-minimo1
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)
desviacion_estandar_1 = stdev(fabrica1)
print("Estadistica descriptiva fabrica 1: \n", "Media: ",media1,
      "\n Mediana: ", mediana1, "\n Moda: ", moda1, "\n Maximo: ", maximo1, "\n Minimo: ", minimo1,
      "\n Rango: ", rango1)
```

```python
maximo2 = max(fabrica2)
minimo2 = min(fabrica2)
rango2 = maximo2-minimo2
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)
desviacion_estandar_2 = stdev(fabrica2)
print("Estadistica descriptiva fabrica 2: \n", "Media: ",media2,
      "\n Mediana: ", mediana2, "\n Moda: ", moda2, "\n Maximo: ", maximo2, "\n Minimo: ", minimo2,
      "\n Rango: ", rango2)
```

```python
from matplotlib.lines import lineStyles
plt.figure(figsize=(8,6))
plt.hist(fabrica1, bins =10, alpha = 0.2, color ='b', edgecolor ="black")
plt.axvline(media1, color = 'b', linestyle = 'dashed', linewidth=2, label =f'Media= {media1: .2f}')
plt.axvline(mediana1, color = 'r', linestyle = 'dashed', linewidth=2, label =f'Mediana= {mediana1: .2f}')
plt.axvline(moda1, color = 'g', linestyle = 'dashed', linewidth=2, label =f'Moda={moda1: .2f}')
plt.axvline(x = desviacion_estandar_1, color ='black', linestyle = 'dashed', linewidth=2, label=f'Devs. Estandar = {desviacion_estandar_1: .2f}')
plt.axvline(x = desviacion_estandar_1, color ='black', linestyle = 'dashed', linewidth=2)

plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Produccion diaria de la Fabrica 1')
plt.legend()
plt.show()
```

```python
from matplotlib.lines import lineStyles
plt.figure(figsize=(8,6))
plt.hist(fabrica2, bins =10, alpha = 0.2, color ='b', edgecolor ="black")
plt.axvline(media2, color = 'b', linestyle = 'dashed', linewidth=2, label =f'Media= {media2: .2f}')
plt.axvline(mediana2, color = 'r', linestyle = 'dashed', linewidth=2, label =f'Mediana= {mediana2: .2f}')
plt.axvline(moda2, color = 'g', linestyle = 'dashed', linewidth=2, label =f'Moda={moda2: .2f}')
plt.axvline(x = desviacion_estandar_2, color ='black', linestyle = 'dashed', linewidth=2, label=f'Devs. Estandar = {desviacion_estandar_2: .2f}')
plt.axvline(x = desviacion_estandar_2, color ='black', linestyle = 'dashed', linewidth=2)

plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Produccion diaria de la Fabrica 1')
plt.legend()
plt.show()
```
