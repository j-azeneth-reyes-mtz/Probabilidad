```python
import numpy as np

# Determinar la cantidad de datos (n)
n = len(autos2['price'])

# Aplicar la regla de Sturges para determinar el número óptimo de bins (k)
k = int(1 + np.log2(n))

print(f"Número de datos: {n}")
print(f"Número óptimo de bins : {k}")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(autos2['price'], bins=k, kde=True)
plt.title('Distribución de Precios de Automóviles')
plt.xlabel('Precio (USD)')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Interpretación del Histograma de Precios

El histograma muestra la distribución de los precios de los automóviles. Observamos lo siguiente:

*   **Distribución Asimétrica Positiva:** La mayor concentración de automóviles se encuentra en el rango de precios más bajos, con una cola larga que se extiende hacia precios más altos. Esto indica que la mayoría de los automóviles son relativamente accesibles, mientras que una menor cantidad tiene precios significativamente más altos.

*   **Rango de Precios Comunes:** La mayoría de los vehículos se encuentran en el rango de precios de aproximadamente \$5,000 a \$15,000, donde la frecuencia es más alta.

*   **Vehículos de Lujo:** Hay una disminución gradual en la frecuencia a medida que los precios aumentan, pero todavía hay un número pequeño de vehículos en los rangos de precios más altos, lo que sugiere la presencia de automóviles de lujo o de gama alta.

*   **Media vs. Mediana:** Dada la asimetría a la derecha, es probable que la media de los precios sea mayor que la mediana, ya que los valores altos "tiran" la media hacia arriba.

En resumen, el mercado de automóviles representado en este dataset está dominado por vehículos de menor y mediano precio, con una proporción menor de automóviles caros o de lujo.

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='body-style', y='price', data=autos2)
plt.title('Distribución de Precios por Tipo de Vehículo (Body Style)')
plt.xlabel('Tipo de Vehículo (Body Style)')
plt.ylabel('Precio (USD)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Interpretación del Diagrama de Caja y Bigotes: Precio vs. Tipo de Vehículo (Body Style)

El diagrama de caja y bigotes nos permite observar la distribución de precios para cada tipo de carrocería, así como identificar la mediana, los cuartiles y la presencia de valores atípicos.

*   **Convertible:** Muestra una distribución de precios relativamente alta, con una mediana significativamente mayor que otros tipos. El rango intercuartílico (IQR) es amplio, lo que indica una variabilidad considerable en los precios, y presenta algunos valores atípicos (puntos por encima del bigote superior), sugiriendo la existencia de modelos muy caros.

*   **Hardtop:** Similar a los convertibles, los hardtops también tienen una mediana de precio alta y un rango de precios que tiende a ser elevado, aunque con menor variabilidad que los convertibles. También se observan valores atípicos de precios altos.

*   **Hatchback:** Presenta una distribución de precios más baja y concentrada. La mediana está en el rango inferior a medio de precios, y el IQR es más estrecho, lo que indica menos variabilidad en comparación con los convertibles y hardtops. Hay algunos valores atípicos de precios más altos, pero la mayoría son más económicos.

*   **Sedan:** Este es el tipo de carrocería más común y muestra una amplia gama de precios. La mediana se encuentra en un punto intermedio, pero el bigote superior y los valores atípicos se extienden a precios muy altos, lo que refleja la existencia de sedanes de lujo o de alta gama junto con modelos más económicos. El IQR es amplio, indicando una gran diversidad de precios.

*   **Wagon:** Los station wagons tienen una distribución de precios en el rango medio a alto, con una mediana superior a la de los hatchbacks y sedanes de gama baja. La variabilidad es moderada y también pueden presentar algunos valores atípicos de precios más altos.

**Conclusiones Generales:**

*   Los **convertibles y hardtops** suelen ser los vehículos más caros, con distribuciones de precios más elevadas.
*   Los **hatchbacks** son, en general, los más económicos, con una distribución de precios más baja y consistente.
*   Los **sedanes** cubren el rango de precios más amplio y son los más diversos en términos de costo, desde opciones económicas hasta vehículos de lujo.
*   La presencia de **valores atípicos** en varios tipos de carrocería (especialmente en convertibles, hardtops y sedanes) sugiere la existencia de modelos premium o de lujo dentro de esas categorías que se desvían significativamente del precio promedio.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='drive-wheels', y='price', data=autos2)
plt.title('Distribución de Precios por Tipo de Tracción (Drive Wheels)')
plt.xlabel('Tipo de Tracción (Drive Wheels)')
plt.ylabel('Precio (USD)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### Interpretación del Diagrama de Caja y Bigotes: Precio vs. Tipo de Tracción (Drive Wheels)

El diagrama de caja y bigotes nos permite analizar la distribución de precios en función del tipo de tracción del vehículo, identificando medianas, rangos y valores atípicos.

*   **Tracción Delantera:** Este tipo de tracción suele asociarse con vehículos de menor a mediano precio. La mediana es la más baja, y el rango intercuartílico indica una concentración de precios en el segmento más económico.

*   **Tracción Trasera:** Los vehículos con tracción trasera tienden a tener una distribución de precios más elevada que los de tracción delantera. La mediana es considerablemente mayor, y el IQR es más amplio, sugiriendo una mayor variabilidad y la presencia de modelos tanto de gama media como de gama alta.

*   **Tracción 4x4:** Los vehículos 4x4 muestran una distribución de precios interesante. Su mediana puede ser similar o ligeramente superior a la de los tracción trasera, pero a menudo presentan un rango de precios más concentrado en el segmento medio a alto. Esto podría indicar que los vehículos 4x4, aunque versátiles, no siempre alcanzan los precios extremos de los modelos de lujo o de alto rendimiento con tracción trasera, pero sí superan consistentemente a los de tracción delantera en precio base.

**Conclusiones Generales:**

*   Los vehículos con **tracción delantera (fwd)** son generalmente los más económicos del dataset.
*   Los vehículos con **tracción trasera (rwd)** son los más caros en promedio, y presentan la mayor dispersión en precios, incluyendo los valores más altos y atípicos, lo que sugiere la inclusión de vehículos de lujo y deportivos.
*   Los vehículos con **tracción 4x4 (4wd)** se posicionan en un rango de precios intermedio a alto, siendo más caros que los de tracción delantera, pero con una variabilidad quizás menor en el extremo superior en comparación con los de tracción trasera.

Presenta un rango intercuartílico (IQR) considerable y una cola superior larga con numerosos valores atípicos que se extienden a precios muy altos. Esto sugiere que, si bien la mayoría de los automóviles con motor frontal son de gama media o económica, también incluye vehículos de lujo o alto rendimiento.

*   **Motor Trasero (Rear):** Solo hay un número muy limitado de vehículos con motor trasero en este dataset. Observamos que la distribución de precios para estos vehículos es significativamente más alta en comparación con los de motor frontal. La mediana de este grupo es mucho mayor, y el rango de precios tiende a estar en el segmento de lujo o alto rendimiento. La menor cantidad de datos para esta categoría puede hacer que la representación sea menos robusta, pero la tendencia general apunta a un mayor costo para los vehículos con motor trasero.
