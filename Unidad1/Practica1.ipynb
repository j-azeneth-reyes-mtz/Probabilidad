```python
#Libreria para agrupar por numros aleatorios
import random
```

```python
fabrica1 = []
for i in range (100):
  numero_aleatorio = round(random.uniform(0,100), 2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range (100):
  numero_aleatorio = round(random.uniform(0,100), 2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
#Libreria de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produccion diaria de la fabrica")
```

```python
plt.hist(fabrica2)
plt.title("Produccion diaria de la fabrica")
```

```python
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print("valores de la fabrica 1")
print("la media es:", media1)
print("la mediana es:", mediana1)
print("la moda es:", moda1)
```

```python
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

print("valores de la fabrica 2")
print("la media es:", media2)
print("la mediana es:", mediana2)
print("la moda es:", moda2)
```

```python
#generar 2 subplot de 1 renglon por 2 columnas para desplegar los 2 histoframas,
# una al lado de la otra
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```
